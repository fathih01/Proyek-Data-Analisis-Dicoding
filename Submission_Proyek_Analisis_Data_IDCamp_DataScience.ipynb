```python
order
```

```python
order.info()
```

```python
order.describe()
```

```python
order['order_status'].unique()
```

```python
# Mengelompokkan data berdasarkan kolom 'order status' dan menghitung jumlah setiap kelompok
statusOrder = order.groupby('order_status').size().reset_index(name='jumlah').sort_values(by='jumlah', ascending=False)
```

```python
statusOrder
```

membagi data menjadi 2 kelompok yakni delivered dan other untuk memudahkan visualisasi

```python
# Buat DataFrame dari data
statusOrder = pd.DataFrame(statusOrder)

# Menghitung jumlah pesanan dengan status 'delivered'
delivered_count = statusOrder[statusOrder['order_status'] == 'delivered']['jumlah'].sum()

# Menghitung jumlah pesanan dengan status 'other'
other_count = statusOrder[statusOrder['order_status'] != 'delivered']['jumlah'].sum()

# Cetak hasilnya
print('Jumlah pesanan dengan status "delivered":', delivered_count)
print('Jumlah pesanan dengan status "other":', other_count)
```

#  Data Visualization

Membuat Visualisasi Data dari 10 daerah dengan transaksi terbesar

```python
# Membuat diagram batang
plt.figure(figsize=(15, 6))
plt.bar(top_10_daerah['customer_city'], top_10_daerah['jumlah'], color='skyblue')
plt.xlabel('Daerah Asal')
plt.ylabel('Jumlah Pelanggan')
plt.title('10 Daerah dengan Jumlah Customer Terbanyak')
# plt.xticks(rotation=45)  # Mengatur label sumbu X agar tidak bertumpuk
plt.tight_layout()

# Menampilkan diagram batang
plt.show()
```

Membuat visualisasi tingkat Keberhasilan Order dengan Pie Chart

```python
# Data jumlah pesanan
labels = ['Delivered', 'Other']
sizes = [96478, 2963]
colors = ['lightblue', 'pink']
explode = (0.1, 0)  # Pemisahan antar potongan pie chart (0.1 artinya memisahkan potongan 'Delivered')

# Membuat pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Persentase Keberhasilan Transaksi')
plt.show()
```

Membuat Visualisasi perbandingan jumlah Review

```python
# Membuat diagram Donut
plt.figure(figsize=(8, 6))
plt.pie(jumlahReview['jumlah'], labels=jumlahReview['review_score'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.gca().add_artist(plt.Circle((0,0),0.70,fc='white'))  # Membuat donut chart
plt.title('Diagram Donut Review Score')
plt.show()
```
